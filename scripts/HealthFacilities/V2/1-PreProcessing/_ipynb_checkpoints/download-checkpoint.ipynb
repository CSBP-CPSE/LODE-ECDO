```python
"""
This script reads in the sources CSV containing the list of open data portals 
Where a direct link to the data is available, it downloads the datasets 
into the raw directory

-Sam Lumley
Dec 2021

"""


import pandas as pd
import requests 

url = 'https://raw.githubusercontent.com/CSBP-CPSE/LODE-ECDO/odhfv2/sources/HealthFacilities/V2/ODHF_v2_Sources.csv'
df = pd.read_csv(url, index_col=0)
```

```python
# filter to provinces Municipal
provinces = df[df['Region type'] == 'Provincial']
provinces = provinces.reset_index()
municipalities = df[df['Region type'] == 'Municipal']
municipalities = municipalities.reset_index()
```

```python
links = df['Direct link'].dropna()
```

```python
links = municipalities['Direct link'].dropna()
```

```python
# Loop through list and download csv files

for index, url in enumerate(links):
    req = requests.get(url)
    url_content = req.content
    
#     province = provinces['Province'][index]
#     filename = provinces['Filename'][index]
#     csv_name = str(province) + '_' + str(filename) + '.csv'
    
#     province = municipalities['Province'][index]
#     municipality = municipalities['Provider/ region'][index]
#     filename = municipalities['Filename'][index]
#     csv_name = str(province) + '_' + str(municipality) + '_' + str(filename) + '.csv'
    
    filename = df['Filename'][index]
    
    csv_name = str(filename) + '.csv'
    csv_file = open(str(csv_name), 'wb')
    csv_file.write(url_content)
    csv_file.close()
```
